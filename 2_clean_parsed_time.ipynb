```python
from datetime import datetime

files=['parsed_time_train1.txt','parsed_time_train2.txt','parsed_time_test.txt']
file_content={}
for file in files:
    print(file)
    with open(file,"rt") as f:
        lines=f.readlines()
    
    new_content=[]
    
    for line in lines:
        line=line.strip()
        frame_id=line.split('\t')[0]
        date_str=line.split('\t')[1]
        
        try:
            parsed_date = datetime.strptime(date_str, '%m-%d-%Y %a %H:%M:%S')
            new_content.append((frame_id,parsed_date))
        except:
            print('except:',line)
            continue
            
    file_content[file]=new_content
            
    
    for i in range(1,len(new_content)-1):
        if new_content[i][1]<new_content[i-1][1] or new_content[i][1]<new_content[i-1][1]:
            print("*********************")
            for j in range(max(0,i-7),min(i+7,len(new_content))):
                print("***" if j==i else "", new_content[j][0],new_content[j][1])
                
    print("dirty line counts:",len(lines))
    print("clean line counts:",len(new_content))
```
